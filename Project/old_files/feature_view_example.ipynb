```python
import hopsworks

project = hopsworks.login()

fs = project.get_feature_store() 
```

```python
air_quality_fg = fs.get_or_create_feature_group(
    name = 'air_quality_fg',
    version = 1
)
weather_fg = fs.get_or_create_feature_group(
    name = 'weather_fg',
    version = 1
)
```

```python
query = air_quality_fg.select_all().join(weather_fg.select_all())
```

```python
query.read()

```

```python
query = air_quality_fg.select_all().join(weather_fg.select_all())

query_show = query.show(5)
col_names = query_show.columns

query_show
```

```python
[t_func.name for t_func in fs.get_transformation_functions()]
```

```python
scal_columns=['pm25','pm10','o3','no2','so2','co','tempmax','tempmin','temp','feelslikemax','feelslikemin','feelslike','dew','humidity','precip','precipprob','precipcover','windgust','windspeed','winddir','sealevelpressure','cloudcover','visibility','solarradiation','solarenergy','uvindex']
```

```python
cat_columns=['conditions']
```

```python
mapping_transformers = {col_name:fs.get_transformation_function(name='standard_scaler') for col_name in col_names if col_name in scal_columns}
```

```python
category_cols = {col_name:fs.get_transformation_function(name='label_encoder') for col_name in col_names if col_name in cat_columns}

mapping_transformers.update(category_cols)
```

```python
feature_view = fs.create_feature_view(
    name = 'air_quality_fv',
    version = 1,
    transformation_functions = mapping_transformers,
    query = query
)
```

```python
feature_view = fs.get_feature_view(
    name = 'air_quality_fv',
    version = 1
)
```

```python

# materialise a training dataset
feature_view.create_train_validation_test_split(
    validation_size = 0.3, 
    test_size = 0.2,
    description = 'air quality training dataset',
    write_options = {"wait_for_job": False},
    data_format = 'csv'
)
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = feature_view.get_train_validation_test_split(training_dataset_version=4)
```
